```python
import dataset_generator
import matplotlib.pyplot as plt
```

```python
def display(score, perf, head, tail) -> None:
    plt.figure(figsize=(16, 9))
    
    plt.subplot(2, 1, 1)
    plt.title("Score")
    plt.imshow(score, cmap="gray", aspect="auto", origin="lower", interpolation="nearest")
    plt.ylabel("MIDI Key")
    plt.xlabel("Frame")
    ylim = len(score)
    plt.ylim([0, ylim])


    plt.subplot(2, 1, 2)
    plt.title("Performance")
    plt.imshow(perf, cmap="gray", aspect="auto", origin="lower", interpolation="nearest")
    plt.ylabel("MIDI Key")
    plt.xlabel("Frame")
    ylim = len(perf)
    plt.vlines(head, 0, ylim, label="head")
    plt.vlines(tail, 0, ylim, label="tail")
    plt.legend()
    plt.ylim([0, ylim])

    plt.show()
```

```python
generator1 = dataset_generator.new(root="../../dataset/newbie-dataset/", shuffle=False, offset_scale=1.000000)
generator2 = dataset_generator.new(root="../../dataset/newbie-dataset/", shuffle=False, offset_scale=0.999999)
```

```python
score, perf, (head, tail) = generator1.__next__()
display(score, perf, head, tail)
```

```python
score, perf, (head, tail) = generator2.__next__()
display(score, perf, head, tail)
```
